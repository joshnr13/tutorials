So, it's a GIS, but it **DOESN'T handle files!**

First, we'll start with a quick illustration of how shapely works and how to visualize things in shapely. Then, we'll work on real data.

```python
import numpy as np #Numerical Python
import matplotlib.pyplot as plt #Math Plotting Library
import shapely.geometry as geom #Shapely, Python Geometries
import descartes # Linker between Shapely and Matplotlib
%matplotlib inline
```

## Fun with circles

We can define a shape and plot it using matplotlib:

```python
circ = geom.Point(5.0, 0.0).buffer(10.0)

f = plt.figure(figsize=(8*1.6, 8))
ax = f.add_subplot(111)
ax.add_patch(descartes.PolygonPatch(circ))
ax.axis('equal')
plt.show()
```

Of course, we can do more than one shape!

```python
circ2 = geom.Point(15.0, 2.0).buffer(5.0)

f = plt.figure(figsize=(8*1.6, 8))
ax = f.add_subplot(111)
ax.add_patch(descartes.PolygonPatch(circ, fc='blue', alpha=.5))
ax.add_patch(descartes.PolygonPatch(circ2, fc='red', alpha=.5))
ax.axis('equal')
plt.show()
```

Now, let's make a bunch of circles randomly in the space!

```python
#np.random.randint?
```

```python
circs = [geom.Point(x,y).buffer(b) for x,y,b in zip(np.random.randint(-20,20,20), np.random.randint(-20,20,20), np.random.randint(1,9,20)) ]

f = plt.figure(figsize=(8*1.6, 8))
ax = f.add_subplot(111)
for c in circs:
    p = np.random.random()
    if p < .33:
        ax.add_patch(descartes.PolygonPatch(c, fc='blue', alpha=.5))
    elif p > .66:
        ax.add_patch(descartes.PolygonPatch(c, fc='red', alpha=.5))
    else:
        ax.add_patch(descartes.PolygonPatch(c, fc='black', alpha=.5))
plt.title('Random Circles!')
ax.axis('equal')
plt.show()
```

We can also do typicaly GIS operations on shapes, like clips, unions, and intersection tests:

```python
s1 = circ.difference(circ2)

print 'Our first red circle intersect our first blue circle?\n', circ.intersects(circ2), '\n\n\n'

print 'Blue NOT Red is:'

f = plt.figure(figsize=(8*1.6, 8))
ax = f.add_subplot(111)
ax.add_patch(descartes.PolygonPatch(s1))
ax.axis('equal')
plt.show()
```

I need this helper function to do some of the ploting we did above. If this is totally foreign to you, review the code or listen to Kelsey's workshop.

```python
def plot_polygon(ax, poly, color='blue', alpha = .5):
    a = np.asarray(poly.exterior)
    ax.add_patch(mplpatch.Polygon(a, facecolor=color, alpha=alpha))
    ax.plot(a[:, 0], a[:, 1], color='black')
```

This is just the beginning of using *one* open source GIS package to do *one* thing. While ArcPy and ArcGIS is very encompassing, it's not the only thing out there. 

Some FOSS-GIS packages are very popular. An old and useful philosophy for software design usually holds in the FOSS-GIS community:

*Do one thing and do it well.*

By stringing together *many* packages (fiona, shapely, matplotlib, descartes here), rather than one monolithic module that has a million relatively undocumented functions (arcpy, arcpy.mapping, ..., etc), we can easily and efficiently do GIS. In the FOSS ecosystem, if your package doesn't do something well, someone else usually will come along and either hack something together that does work well or, using your own code, *make* a solution and give it back to you! 
Since this is all open source, you can see the code and understand how it works. For libraries like Shapely, that are BSD or MIT licensed, you can even construct your own **for-profit applications** around them, if that suits your interest.

But, the Free and Open Source computing community is built on *two* ideas of the word 'free.' *Gratis*, or 'free as in beer', is the one y'all are probably thinking. Since these are all available to download at no cost, they're "free software." But, I also mean "free" in another sense: **no one company, author, or agency can make these tools unavailable for the public to use.** That is, these tools are *libre* software, or liberated/emancipated from the restrictions that hold other kinds of propertary software. As a GIS professional, this means you always will have a set of open and free tools embedded in an open, social, and public community of users and developers, all with a vested interest in making the software better.
